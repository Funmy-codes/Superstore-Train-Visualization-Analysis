```python
# I imported the necessary libraries and the dataset I will be working with
import pandas as pd
import plotly.express as pe
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.graph_objects as pg
pd.set_option('display.max_columns', None)
data = pd.read_csv('/content/superstore_train.csv')
data.head(4)
```

```python
# I seperated the categorical data from the numerical data
categorical = data.select_dtypes(include = ['object', 'category'])   #---- for the categorical data
numerical = data.select_dtypes(include = 'number')     ##---- for the numerical data
print(f'Categorical data')
display(categorical.head(4))
print(f'\n\nNumerical data')
display(numerical.head(2))
```

```python
# I did some visualisation showimg the segments of sales
sel = data[['Segment', 'Sales']]    ### I selected the columns I will be working with
sel = sel.groupby('Segment')[['Sales']].mean().sort_values(by = 'Sales', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Sales', color = 'Sales', width = 700, height=500, title='Distrbution of segment')     #---- I brought in the visualization I will be working with
fig.show()
```

```python
# The visualization was on the total quantity ordered for all consumer and corporate products by each state
sel = data[['State', 'Quantity']]          ### I selected the columns I will be working with
sel = sel.groupby('State')[['Quantity']].mean().sort_values(by = 'Quantity', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Quantity', color = 'Quantity', width = 700, height=500, title='Distrbution of State')       #---- I brought in the visualization I will be working with
fig.show()
```

```python
# I worked on the visualization of the city that sells the most Consumer Product
sel = data[['City', 'Quantity']]
sel = sel.groupby('City')[['Quantity']].mean().sort_values(by = 'Quantity', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Quantity', color = 'Quantity', width = 700, height=500, title='Distrbution of City')
fig.show()
```

```python
# The visualization of the total sales made on Chairs in each state
sel = data[['State', 'Sub-Category', 'Sales']]
sel = sel.loc[(data['Sub-Category'] == 'Chairs')]
sel = sel.groupby('State')[['Sales']].mean().sort_values(by = 'Sales', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Sales', color = 'Sales', width = 700, height=500, title='Distrbution of State')
fig.show()
```

```python
# I checked to see if there is a relationship between sales and Profit
sel = data[['Sales', 'Profit']]
fig = pe.scatter(data_frame= sel, x = 'Sales', y = 'Profit', title= f"There exist a {sel['Profit'].corr(sel['Sales']).round(2)} % Correlation between the Sales and Profit")
fig.show()
```

```python
# Visualization on the city whose company have the most orders.
sel = data[['City', 'Quantity']]
sel = sel.groupby('City')[['Quantity']].mean().sort_values(by = 'Quantity', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Quantity', color = 'Quantity', width = 700, height=500, title='Distrbution of City')
fig.show()
```

```python
# I worked on the visualization of the product category that is most ordered in Vacaville
sel = data[['Category', 'City', 'Quantity']]
sel = sel.loc[(data['City'] == 'Vacaville')]
sel = sel.groupby('Category')[['Quantity']].mean().sort_values(by = 'Quantity', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Quantity', color = 'Quantity', width = 700, height=500, title='Distrbution of Product Category')
fig.show()
```

```python
# I visualized the distribution of shipping mode for each Product Segment
sel = data[['Ship Mode', 'Quantity']]
sel = sel.groupby('Ship Mode')[['Quantity']].mean().sort_values(by = 'Quantity', ascending = False)
sel

fig = pe.bar(data_frame= sel, x = sel.index, y= 'Quantity', color = 'Quantity', width = 700, height=500, title='Distrbution of Ship Mode')
fig.show()
```
